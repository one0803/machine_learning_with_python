```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
```

```python
#CSVデータの読み込み
anime = pd.read_csv('anime.csv')
ratings = pd.read_csv('rating.csv')
```

```python
#評価値をヒストグラムにして表示
ratings['rating'].hist(bins=11)
plt.show()
```

```python
#リコメンドの精度を上げるために、人気のアニメのみを取り扱う
anime = anime[anime['members'] > 10000]
```

```python
#欠損値の確認
anime.isnull().sum()
```

```python
#欠損値の除去
anime = anime.dropna()
```

```python
#評価値-1を除去し、除去されていることを確認
ratings = ratings[ratings['rating'] >= 0]
round(ratings.describe(), 2)
```

```python
#animeとratingsをマージしたデータフレームをつくる suffixesは重複する列の接尾辞を指定する
mergeddf = ratings.merge(anime, left_on='anime_id', right_on='anime_id', suffixes=['_user', ''])
mergeddf.head()
```

```python
round(mergeddf.describe(), 2)
```

```python
#今回使うのはuser_id, name, rating_user
mergeddf = mergeddf[['user_id', 'name', 'rating_user']]
#重複の除去(必要？)
mergeddf = mergeddf.drop_duplicates(['user_id', 'name'])
mergeddf.head()
```

```python
#疎行列の作成
anime_pivot = mergeddf.pivot(index='name', columns='user_id', values='rating_user').fillna(0)
anime_pivot.head()
```

```python
#疎行列の圧縮
anime_pivot_sparse = csr_matrix(anime_pivot)
```

```python
#knnで学習
knn = NearestNeighbors(n_neighbors=11, algorithm='brute', metric='cosine')
model_knn = knn.fit(anime_pivot_sparse)
```

```python
#文字検索機能
def searchanime(string):
    print(anime_pivot[anime_pivot.index.str.contains(string)].index[0:])

searchanime('Hajime')
```

```python
model_knn.kneighbors
```

```python
Anime = 'Hajime no Ippo'

anime_pivot.iloc[anime_pivot.index==Anime]
```

```python
distance, indices = model_knn.kneighbors(anime_pivot.iloc[anime_pivot.index==Anime].values.reshape(1, -1))
```

```python
#はじめの一歩を見た人に対するおすすめ
print('Recommendations if you like the anime {0}:\n'.format(Anime))
for i in range(len(distance.flatten())):
    if i != 0:
        print('{0}: {1} with distance: {2}'.format(i, anime_pivot.index[indices.flatten()[i]], distance.flatten()[i]))
```

```python
#君の名はを見た人に対するおすすめ
Anime = 'Kimi no Na wa.'
distance, indices = model_knn.kneighbors(anime_pivot.iloc[anime_pivot.index==Anime].values.reshape(1, -1))

print('Recommendations if you like the anime {0}:\n'.format(Anime))
for i in range(len(distance.flatten())):
    if i != 0:
        print('{0}: {1} with distance: {2}'.format(i, anime_pivot.index[indices.flatten()[i]], distance.flatten()[i]))
```
